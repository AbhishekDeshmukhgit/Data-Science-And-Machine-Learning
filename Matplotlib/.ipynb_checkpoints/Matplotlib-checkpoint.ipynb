# Matplotlib workflow example

```python
# 1.Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2.Setput plot
fix,ax=plt.subplots(figsize=(5,10))

# 3.Plot the data
ax.plot(x,y)

# 4.Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")
# Save and show the plot
fig.savefig("sample.png")
```

# Two options for subplots

```python
# Subplot option1 
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                     ncols=2,
                                     figsize=(10,5))
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(butterprices.keys(),butterprices.values())
ax4.hist(np.random.random(10))
```

```python
#Subplot option 2
fig,ax=plt.subplots(nrows=2,
                    ncols=2,
                   figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(butterprices.keys(),butterprices.values());
ax[1,1].hist(np.random.random(10));
```

```python
heart=pd.read_csv("heart-disease.csv");
heart.head()
```

```python
# Create a histogram of age
heart["age"].plot.hist()
```

```python
heart.plot.hist(figsize=(10,30),subplots="True")
```

```python
over_50=heart[heart["age"]>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind='scatter',
            x='age',
             y='chol',
            c='target')
```

```python
#OO Method
fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
```

```python
# OO Method from scratch
fig,ax=plt.subplots(figsize=(10,6))

#Plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])

# Customize 
ax.set(title="Heart Disease",
      xlabel="Age",
      ylabel="Cholestrol")

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                           figsize=(10,10),
                           sharex=True)

# Add data to the ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")

# Customize the ax0
ax0.set(title="Heart Disease and cholestrol levels",
       ylabel="Cholestrol")

#Change the x axis limits
ax0.set_xlim([50,80])

# Add a legend to a cholestrol
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")


# Add data to the ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="PuBu")

# Customize the ax01
ax1.set(title="Heart Disease and cholestrol levels",
       xlabel="Age",
       ylabel="thalach")

# Change ax1 x axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend to a cholestrol
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")


# ADD title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
